Armarmos el heatmap con hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)

```python
# Calculando la matriz de correlación solo con variables numéricas
CorrelacionVariables2 = df_train.corr(numeric_only=True)


# Creando el mapa de calor
f, ax = plt.subplots(figsize=(40, 36))
sns.heatmap(CorrelacionVariables2, vmax=.8, square=True, annot=True, fmt='.2f', cmap='coolwarm')

plt.show()
```

```python
# Calculando la matriz de correlación solo con variables numéricas
CorrelacionVariables2 = df_train.corr(numeric_only=True)

# Reemplazamos los valores de correlación menores o iguales a 0.75 con NaN
CorrelacionVariablesFiltradas = CorrelacionVariables2.where(CorrelacionVariables2 > 0.75)

# Creando el mapa de calor con las correlaciones filtradas
f, ax = plt.subplots(figsize=(30, 28))
sns.heatmap(CorrelacionVariablesFiltradas, vmax=1, square=True, annot=True, fmt='.2f', cmap='coolwarm')

plt.show()
```

```python
# Calculando la matriz de correlación solo con variables numéricas
CorrelacionVariables2 = df_train.corr(numeric_only=True)

# Filtrando la matriz de correlación para valores superiores a 0.75
CorrelacionesAltas = CorrelacionVariables2.where(CorrelacionVariables2 > 0.75)

# Asignar la matriz filtrada a una variable
CorrelacionesSignificativas = CorrelacionesAltas

# Mostrar la matriz filtrada
CorrelacionesSignificativas
```

### 4.3.- Scatter Plots entre Sale Price y variables Correlacionados
